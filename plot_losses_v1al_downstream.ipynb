```python
import os
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

def get_losses(event_file):
    event_acc = EventAccumulator(event_file)
    event_acc.Reload()
    loss_data = event_acc.Scalars('Loss/test_cls')
    loss_steps = [entry.step for entry in loss_data]
    loss_values = [entry.value for entry in loss_data]
    return loss_steps, loss_values

def plot_losses(tb_event_files):
    plt.figure(figsize=(12, 6))
    for label, seed_files in tb_event_files.items():
        for seed, event_file in seed_files.items():
            steps, losses = get_losses(event_file)
            plt.plot(steps, losses, label=f"{label} (Seed {seed})")
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Losses vs Steps')
    plt.show()

plot_losses(tb_event_files)
```

```python
import numpy as np

def merge_losses(tb_event_files):
    merged_data = {}
    for label, seed_files in tb_event_files.items():
        merged_steps = []
        merged_losses = []
        for seed, event_file in seed_files.items():
            steps, losses = get_losses(event_file)
            merged_steps.append(steps)
            merged_losses.append(losses)
        merged_data[label] = {'steps': merged_steps, 'losses': merged_losses}
    return merged_data

def mean_and_variance(data):
    max_len = max([len(steps) for steps in data['steps']])
    steps = np.zeros(max_len)
    losses = np.zeros((len(data['losses']), max_len))
    
    for i, (seed_steps, seed_losses) in enumerate(zip(data['steps'], data['losses'])):
        steps[:len(seed_steps)] = seed_steps
        losses[i, :len(seed_losses)] = seed_losses
        losses[i, len(seed_losses):] = seed_losses[-1]

    mean_losses = np.mean(losses, axis=0)
    var_losses = np.std(losses, axis=0)
    
    return steps, mean_losses, var_losses

def plot_mean_and_variance(tb_event_files):
    plt.figure(figsize=(12, 6))
    merged_data = merge_losses(tb_event_files)
    
    for label, data in merged_data.items():
        steps, mean_losses, var_losses = mean_and_variance(data)
        plt.plot(steps, mean_losses, label=f"{label}")
        plt.fill_between(steps, mean_losses - var_losses, mean_losses + var_losses, alpha=0.2)
    
    plt.xlabel('Steps')
    plt.ylabel('Loss, Mouse Binary Classification')
    plt.legend()
    # plt.title('Mean and Variance of Losses vs Steps')
    plt.show()

def moving_average(data, window_size):
    cumsum = np.cumsum(np.insert(data, 0, 0))
    smoothed_data = (cumsum[window_size:] - cumsum[:-window_size]) / window_size
    return smoothed_data

def plot_smoothed_mean_and_variance(tb_event_files, window_size=10):
    plt.figure(figsize=(12, 6))
    merged_data = merge_losses(tb_event_files)
    
    for label, data in merged_data.items():
        steps, mean_losses, var_losses = mean_and_variance(data)
        smoothed_mean_losses = moving_average(mean_losses, window_size)
        smoothed_var_losses = moving_average(var_losses, window_size)
        smoothed_steps = steps[window_size - 1:]
        
        plt.plot(smoothed_steps, smoothed_mean_losses, label=f"{label}")
        plt.fill_between(smoothed_steps, smoothed_mean_losses - smoothed_var_losses, smoothed_mean_losses + smoothed_var_losses, alpha=0.2)
    
    plt.xlabel('Steps')
    plt.ylabel('Loss, Mouse Binary Classification')
    plt.legend()
    plt.title(f'Smoothed Mean and Variance of Losses vs Steps (Window Size: {window_size})')
    plt.show()

plot_smoothed_mean_and_variance(tb_event_files, window_size=10)
```
